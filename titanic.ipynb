```python
train = pd.read_csv('train.csv')
test  = pd.read_csv('test.csv')
```

```python
X_train = train.iloc[:,2:]
Y_train = train.iloc[:,1]
X_test = test.iloc[:,2:]
Y_test = test.iloc[:,1]
X_train.head(10)
```

```python
np.shape(X_train)
```

```python
X_train.isna().sum()
```

```python
X_train_sub = X_train[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Embarked']]
X_test_sub = X_test[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Embarked']]

X_train_sub.head(10)
```

```python
dummies0 = pd.get_dummies(X_train_sub['Pclass'])
dummies0.head()
```

```python
dummies1 = pd.get_dummies(X_train_sub['Sex'])
dummies1.head()
```

```python
dummies2 = pd.get_dummies(X_train_sub['Embarked'])
dummies2.head()
```

```python
X_train_sub_dum = pd.concat([X_train_sub, dummies0, dummies1, dummies2], axis = 1)
X_train_sub_dum
```

```python
X_train_sub_dum = X_train_sub_dum[['Age', 'SibSp', 'Parch', 'Fare', 'Lower', 'Middle', 'Upper', 'female', 'male', 'Cherbourg', 'Queenstown', 'Southampton']]
X_train_sub_dum
```

```python
dummies0_t = pd.get_dummies(X_test_sub['Pclass'])
dummies1_t = pd.get_dummies(X_test_sub['Sex'])
dummies2_t = pd.get_dummies(X_test_sub['Embarked'])
X_test_sub_dum = pd.concat([X_test_sub, dummies0_t, dummies1_t, dummies2_t], axis = 1)
X_test_sub_dum = X_test_sub_dum[['Age', 'SibSp', 'Parch', 'Fare', 'Lower', 'Middle', 'Upper', 'female', 'male', 'Cherbourg', 'Queenstown', 'Southampton']]
```

```python
X_train_sub_dum.describe()
```

```python
X_train_sub_dum['Age'] = X_train_sub_dum['Age'].fillna(X_train_sub_dum['Age'].mean())
X_test_sub_dum['Age']  = X_test_sub_dum['Age'].fillna( X_train_sub_dum['Age'].mean())
X_train_sub_dum.describe()
```

```python
X_train_sub_dum
```

```python
for col in ['Age', 'SibSp', 'Parch', 'Fare']:
    mean = X_train_sub_dum[col].mean()
    std  = X_train_sub_dum[col].std()
    X_train_sub_dum[col] =  (X_train_sub_dum[col] - mean)/std
    X_test_sub_dum[col]  =  (X_test_sub_dum[col] - mean)/std
    
X_train_sub_dum
```

```python
#modelo funcional
from tensorflow import keras
import tensorflow as tf
input   = keras.layers.Input(shape=X_train_sub_dum.shape[1])
hidden1 = keras.layers.Dense(6, activation='relu')(input)
hidden2 = keras.layers.Dense(3, activation='relu')(hidden1)
output  = keras.layers.Dense(1, activation='sigmoid')(hidden2)
model   = keras.models.Model(inputs=input, outputs=output)
```

```python
lr = 0.001
batch_size = 150
n_epochs = 150
optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer = optimizer,
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train_sub_dum,
                 Y_train,
                 batch_size = batch_size,
                 epochs=n_epochs,
                 verbose=1,
                 shuffle=True,
                 validation_data=(X_test_sub_dum, Y_test))
```
